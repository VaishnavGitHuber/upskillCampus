### About the dataset
* Crop: This denotes the type of crop being cultivated. For example, it could be wheat, rice, maize, soybeans, etc.
* State: Refers to the geographical location or administrative division where the crop is being cultivated. Different states may have varying agro-climatic conditions, soil types, water availability, and government policies, all of which can influence the cost and productivity of cultivation.
* Cost of Cultivation (₹/Hectare) A2+FL: This represents the total cost incurred per hectare of land during cultivation. The A2+FL cost includes all paid-out expenses, such as seeds, fertilizers, pesticides, hired labor, fuel, etc. It does not include the imputed value of family labor. This cost category is used for calculating the Minimum Support Price (MSP) for crops in India.
* Cost of Cultivation (₹/Hectare) C2: This represents the comprehensive cost of cultivation per hectare of land. It includes A2+FL costs along with imputed costs of family labor, rent paid for leased-in land, interest on owned land and fixed capital assets, and other costs of capital, like depreciation on farm machinery and buildings. The C2 cost is considered more comprehensive than A2+FL and is used for assessing the true economic cost of cultivation.
* Cost of Production (₹/Quintal) C2: This denotes the cost incurred to produce one quintal (100 kg) of the crop. It's derived from the total cost of cultivation (C2) divided by the yield per hectare. This metric helps in understanding the cost efficiency of crop production.
* Yield (Quintal/Hectare): This indicates the productivity of the crop, measured as the quantity harvested per hectare of land. It's expressed in quintals per hectare. Higher yield signifies better productivity and efficiency in cultivation practices.

```python
# Agregation for understanding the loss and profit with production and yeild
df.groupby(["Crop","State"])["Cost of Production (`/Quintal) C2","Yield (Quintal/ Hectare) "].sum()
```

* Quintal: A quintal is a unit of weight equal to 100 kilograms (kg). It is commonly used in agriculture to measure the quantity of crops, particularly grains and other agricultural products.
* Hectare: A hectare is a unit of area commonly used in agriculture, forestry, and land measurement. It is equal to 10,000 square meters or approximately 2.47 acres. Hectares are often used to measure the size of agricultural fields or land parcels.

```python
# Perform groupby operation and aggregate using sum()
grouped_data = df.groupby(["Crop", "State"])[["Cost of Production (`/Quintal) C2", "Yield (Quintal/ Hectare) "]].sum()

# Reset index to make the groupby columns regular columns
grouped_data_reset = grouped_data.reset_index()

# Plot the data for arhar
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
for crop in ["ARHAR"]:
    crop_data = grouped_data_reset[grouped_data_reset["Crop"] == crop]
    plt.plot(crop_data["State"], crop_data["Cost of Production (`/Quintal) C2"], label=f"{crop} - Cost of Production",marker="o")
    #plt.plot(crop_data["State"], crop_data["Yield (Quintal/ Hectare) "], label=f"{crop} - Yield")
plt.xlabel("State")
plt.ylabel("Value")
plt.title("ARHAR:Cost of Production by Crop and State")
plt.legend()
plt.grid(True)

plt.subplot(2,1,2)
for crop in ["ARHAR"]:
    crop_data = grouped_data_reset[grouped_data_reset["Crop"] == crop]
    #plt.plot(crop_data["State"], crop_data["Cost of Production (`/Quintal) C2"], label=f"{crop} - Cost of Production")
    plt.plot(crop_data["State"], crop_data["Yield (Quintal/ Hectare) "], label=f"{crop} - Yield",color="orange",marker="o")
plt.xlabel("State")
plt.ylabel("Value")
plt.title("ARHAR:yeild by Crop and State")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

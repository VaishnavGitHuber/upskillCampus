```python
# importing necessorty libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

# Ignore incompactable warnings 
import warnings
warnings.filterwarnings(action="ignore")
```

### About the dataset:
* Crop: Specifies the type of crop being analyzed, such as Total Foodgrains, Rice, Wheat, Jowar, Bajra, etc.
* Production (2006-07) to Production (2010-11): These columns represent the production of the respective crop in the specified year. The production values are likely given in units such as million metric tons or thousand metric tons.
* Area (2006-07) to Area (2010-11): These columns indicate the area of land harvested for the respective crop in the specified year. The area values are likely given in units such as million hectares or thousand hectares.
* Yield (2006-07) to Yield (2010-11): These columns represent the yield of the respective crop in the specified year. Yield is typically measured as the quantity of crop produced per unit area of land (e.g., kilograms per hectare or quintals per hectare).

```python
# Loading dataset
df = pd.read_csv("datafile (2).csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# Separating the dataframe for the easy analysis
production_df = df.iloc[:,1:6]
area_df = df.iloc[:,6:11]
yield_df = df.iloc[:,11:]
```

```python
# Picturaising for the bettwe understanding of the prduction,area filed and yield
c = 1

plt.figure(figsize=(25,25))
#plt.suptitle("Area VS Yeild Analysis",color="r",size=5)
for i in range(df.shape[0]):
    plt.subplot(11,5,c)
    plt.title(df["Crop             "].iloc[i] + " Analysis")
    c += 1
    plt.plot([2006,2007,2008,2009,2010],production_df.iloc[i],color="orange",marker="*",label="Production")
    plt.plot([2006,2007,2008,2009,2010],area_df.iloc[i],color="r",marker="o",label="Area")
    plt.plot([2006,2007,2008,2009,2010],yield_df.iloc[i],color="k",marker="+",label="Yield")
    plt.xlabel("Year")
    plt.grid(True)
    plt.legend()
plt.tight_layout()
plt.show()
    
```
